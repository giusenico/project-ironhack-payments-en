# Data Cleaning

```python
# List of columns to convert to datetime format
date_columns = [
    'created_at', 'updated_at', 'paid_at', 
    'from_date', 'to_date'
]

# Convert all columns to datetime format.
for col in date_columns:
    fees_df[col] = pd.to_datetime(fees_df[col], errors='coerce')
```

```python
fees_df.info()
```

```python
# I display the number of null values for each column
fees_df.isnull().sum()
```

```python
# Filter the DataFrame to keep only rows where 'cash_request_id' is not null
fees_df = fees_df[fees_df['cash_request_id'].notnull()]

fees_df.isnull().sum()
```

```python
# Replace null values in the category column
fees_df['category'] = fees_df['category'].fillna("No Incident")
fees_df.sample(3)
```

```python
fees_df.isnull().sum()
```

### Revenue Generated by the Cohort

```python
# Create the column revenue_month by extracting the year and month from the column 'created_at'
fees_df['revenue_month'] = fees_df['created_at'].dt.to_period('M')
```

```python
fees_df.sample(5)
```

```python
# Calculate the total_mount for each revenue_month
total_amount = fees_df.groupby('revenue_month')['total_amount'].sum()
total_amount
```

```python
# chart creation for total_amount

# Convert index to strings
total_amount.index = total_amount.index.astype(str)

plt.figure(figsize=(12, 6))
plt.plot(total_amount.index, total_amount.values, marker='o')

plt.title('Total Revenue per Month')
plt.xlabel('Revenue Month')
plt.ylabel('Total Amount')

plt.xticks(rotation=45)

plt.show()
```

The gradual increase in revenue through October, followed by a sharp decline, is in line with trends we had noted in the cash requests analysis

### Combined comparison of Frequency of Service Use and Revenue Generated by the Cohort

```python
freq_use = pd.read_csv("frequency_use.csv")

freq_use
```

```python
# Convert index to strings for graph
total_amount.index = total_amount.index.astype(str)
freq_use['cohort_month'] = freq_use['cohort_month'].astype(str)

# Creating the combined graph
fig, ax1 = plt.subplots(figsize=(14, 7))

# Revenue graph
ax1.plot(total_amount.index, total_amount.values, color='b', marker='o', label='Total Revenue')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Revenue', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Graph for frequency of use
ax2 = ax1.twinx()
ax2.plot(freq_use['cohort_month'], freq_use['user_id'], color='g', marker='o', label='Frequency of Use')
ax2.set_ylabel('Frequency of Use', color='g')
ax2.tick_params(axis='y', labelcolor='g')

plt.title('Total Revenue and Frequency of Use Over Time')
fig.tight_layout()
plt.show()
```

The graph clearly shows the trend in revenue and frequency of use over time. A similar trend can be seen between the two:

- Both grow steadily until October 2020.
- In October, both reach a significant peak.
- Thereafter, both drop dramatically in November 2020.
